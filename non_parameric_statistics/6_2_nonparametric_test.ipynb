```r
x=mtcars[,1]
y=mtcars[,6]
M=cbind(x,y)
M=M[-which(duplicated(x)),]
M=M[-which(duplicated(M[,2])),]
```

```r
spearman.test(M[,1], M[,2], approximation = "exact")
```

```r
spearman.test(M[,1], M[,2], approximation = "t-distribution")
```

```r
spearman.test(M[,1], M[,2], approximation = "AS89")
```

```r
N=3
F=function(x){
    pspearman(x,N, approximation = "exact")
}
```

```r
maxval=(N^3-N)/3
```

```r
a=seq(from=0, to=maxval, length=1000)
```

```r
plot(a, sapply(a,F), type='l')
```

```r
x <- c(1.83,  0.50,  1.62,  2.48, 1.68, 1.88, 1.55, 3.06, 1.30)
y <- c(0.878, 0.647, 0.598, 2.05, 1.06, 1.29, 1.06, 3.14, 1.29)
wilcox.test(x, y, paired = TRUE)
wilcox.test(x, y, paired = TRUE, alternative = "greater")
wilcox.test(x, y, paired = TRUE, alternative = "less")
```

```r
wilcox.test(x - y)  
wilcox.test(x - y, alternative = "greater")    
wilcox.test(x - y, alternative = "less")   
```

```r
wilcox.test(y - x, alternative = "less",
            exact = FALSE, correct = FALSE) # H&W large sample
                                            # approximation
```
